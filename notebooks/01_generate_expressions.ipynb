```python
import pandas as pd
MAXIT = 10
```

```python
statements = set('S')
operators = ['T','N']
```

```python
for iteration in range(MAXIT):
    new_expressions = []
    for statement in statements:
        for operator in operators:
            new_expressions.append(operator+statement)
    for expression in new_expressions:
        statements.add(expression)

df = (
    pd.DataFrame({'expression':sorted(list(statements))})
    .assign(expression_length=lambda df: df['expression'].str.len())
    .sort_values(['expression_length', 'expression'])
    .reset_index()
    .drop(columns=['index'])
)

df
```

```python
# check data frame 
display(df.head(4))
display(df[df['expression_length']==8])
display(df.tail(5))
```

```python
df.to_csv('../data/expressions_df.csv', index=False)
```

```python
sentences = set('S')
logical_components = ['T','N']
```

```python
for iteration in range(MAXIT):
    new_sentences = []
    for sentence in sentences:
        new_sentences.append(logical_components[0]+sentence)
        new_sentences.append(logical_components[1]+sentence)
    for sentence in new_sentences:
        sentences.add(sentence)
df =(
    pd.DataFrame({'sentence':sorted(list(sentences))})
    .assign(sentence_length=lambda df: df['sentence'].str.len())
    .sort_values(['sentence_length', 'sentence'])
    .reset_index()
    .drop(columns=['index'])
)

display(df.head(9))
display(df[df['sentence_length']==8])
```
